Imports and method definition

```python
import PIL
from PIL import Image
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from PIL import Image


def openJPG(filename):
  return Image.open(filename)

def saveJPG(img, filename):
  img.save(filename, "JPEG", quality=80)

def imgtoarray(img):
  return np.array(np.uint8((img)))

def arraytoimg(img):
  return Image.fromarray(img)

def invertChannel(img, channel):
  #img as nparray
  img[:,:,channel] = 255.0 - img[:,:,channel]
  
def scaleChannel(img, channel, value):
  #img as nparray
  img[:,:,channel] = img[:,:,channel] * value
```

Interactive visualisation:

```python
#@title Interactive visualisation. { run: "auto" }
def plot(filepath, rv=1, gv=1, bv=1, ri=False, gi=False, bi=False, save=False):
  img = openJPG(filepath)
  npimg=imgtoarray(img)
  scaleChannel(npimg, 0, rv)
  scaleChannel(npimg, 1, gv)
  scaleChannel(npimg, 2, bv)
  if ri:
    invertChannel(npimg, 0)
  if gi:
    invertChannel(npimg, 1)
  if bi:
    invertChannel(npimg, 2)

  if save:
    saveJPG(arraytoimg(npimg), "/content/modified.jpg")

  plt.figure()
  plt.title("Your image")
  plt.imshow(npimg)
  plt.show()



filepath = "luke.jpg" #@param {type:"string"}
scale_red = 1 #@param {type:"slider", min:0, max:1, step:0.1}
scale_green = 1 #@param {type:"slider", min:0, max:1, step:0.1}
scale_blue = 1 #@param {type:"slider", min:0, max:1, step:0.1}
invert_red = False #@param {type:"boolean"}
invert_green = False #@param {type:"boolean"}
invert_blue = False #@param {type:"boolean"}
save = False #@param {type:"boolean"}
plot(filepath, scale_red, scale_green, scale_blue, invert_red, invert_green, invert_blue, save)
```
